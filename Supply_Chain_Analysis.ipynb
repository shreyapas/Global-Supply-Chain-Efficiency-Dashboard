Converted the “Shipping date (DateOrders)” column to a proper datetime format using pd.to_datetime() with infer_datetime_format=True. infer_datetime_format=True helps Pandas automatically detect different formats.

errors='coerce' converts any unreadable value into NaT (null datetime) safely

```python
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'], errors='coerce', infer_datetime_format=True)
```

```python
df['shipping date (DateOrders)'] = df['shipping date (DateOrders)'].dt.date
```

**Step - 3: Data Cleaning and Transformation**

```python
df=df.drop(['Customer Email'],axis=1)
```

```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
```

```python
(df[numeric_cols] < 0).sum()
```

```python
df['Order_Year'] = pd.to_datetime(df['order date (DateOrders)'], errors='coerce').dt.year
df['Order_Month'] = pd.to_datetime(df['order date (DateOrders)'], errors='coerce').dt.month
```

```python
df.to_csv("/content/Supply Chain Dataset.csv", index=False)
```
